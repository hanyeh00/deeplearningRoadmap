#tip


### train dev set and test should be from the same distribution

#What is train/dev/test split
-It is important to choose the dev and test sets from <font color="#00Ab">**the same distribution**</font>  and it must be <font color="red">**taken randomly**</font> from all the data.

## train_set:
-learning algorithms use training data with the **objective** of performing really well on this *training* data but **our goal** is a <font color="#00FFَF">**generalization**</font> to *unseen* data than performing well on training data.
- اگر کل داده ها را روی داده های آموزش استفاده کنیم و دقت 99 درصد داشته باشیم در واقع داریم غیر مستقیم میگیم <font color="#01BFFF">یک مدلی ساختم که روی داده های داده شده دقت 99 درصد دارد ولی روی داده های ناشناخته(تست) مطمین نیستم**</font>


## dev_set:


- The goal of **dev-set** is to rank the models in term of their accuracy and helps us decide which model to proceed further with

- *Choices* of **parameters** are made by learning algorithm and *choice* of **model** is made by us by intentionally picking the best model performing well on dev-set.


-If we don’t have dev data then we’ll train all the models and pick the model with the best performance on training data


## test_st:
We use **test-set** as a proxy for unseen data and evaluate our model on test-set.
